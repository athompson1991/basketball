```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

# Load data
data_directory = "~/Desktop/Research/dataset"
games = pd.read_csv(data_directory + "/games.csv")
boxscore = pd.read_csv(data_directory + "/boxscore.csv")
money_line = pd.read_csv(data_directory + "/money_line.csv")

games = pd.DataFrame(games)
boxscore = pd.DataFrame(boxscore)
money_line = pd.DataFrame(money_line)

# Prepare games
games["game_date"] = pd.to_datetime(games["game_date"])

# Prepare moneyline
money_line = money_line[money_line["home"] != "Off"]
money_line["date_url"] = pd.to_datetime(money_line["date_url"])
money_line["home"] = money_line["home"].astype(float)
money_line["away"] = money_line["away"].astype(float)

# Prepare boxscore
# 1. Replace all NaN elements with 0s.
boxscore = boxscore.fillna(0)

# 2. Convert datatypes to float
boxscore["fg_pct"] = boxscore["fg_pct"].astype(float)
boxscore["fg3_pct"] = boxscore["fg3_pct"].astype(float)
boxscore["ft_pct"] = boxscore["ft_pct"].astype(float)

# test
boxscore_join = boxscore[["code", "team", "player_code", "fg_pct", "fg3_pct", "ft_pct"]]
games_join = games[["code", "game_date", "home_code", "visiting_code", "home_team", "visiting_team", "winner"]]
df = pd.merge(games_join, boxscore_join, how = 'inner', on=['code', 'code'])

df["home_wins"] = np.where(df["winner"] == df["home_code"], 1, 0)
#boxscore = boxscore.groupby(["code", "team"])['player_code'].apply(' '.join).reset_index()

#df.shape
df.head(n=20)

df.pivot(index="code", column="player_code", values="fg_pct")
```
